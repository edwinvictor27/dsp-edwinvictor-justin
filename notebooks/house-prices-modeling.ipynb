# **DATA PRE-PROCESSING for the purpose of training the model**

```python
selected_columns = ['LotArea', 'GrLivArea', 'Neighborhood', 'HouseStyle'] # 2 continuos and 2 Categorical
target_column = ['SalePrice'] #Output column
```

```python
Columns_selected = Training_dataset[selected_columns]
Columns_selected
```

```python
Columns_selected.isnull().sum() #checking if there's any null value
```

```python
target_attribute = Training_dataset[target_column]
target_attribute
```

```python
target_attribute.isnull().sum() #checking if there's any null values
```

```python
continuos_datatype_col = ['LotArea', 'GrLivArea']
discrete_datatype_col = ['Neighborhood', 'HouseStyle']
```

```python
merged_attributes = pd.concat([Columns_selected,target_attribute], axis=1)
merged_attributes #concatenating selected features together as single dataframe
```

```python
Columns_selected['Neighborhood'].nunique() #to get the count of different values from the categorical data type column
```

```python
Columns_selected['HouseStyle'].nunique() #to get the count of different values from the categorical data type column
```

```python
implementing_one_hot_encode_Neighborhood = pd.get_dummies(merged_attributes[discrete_datatype_col])
implementing_one_hot_encode_Neighborhood.head() #performing one hot encoding for the selected discrete(categorical) features
```

# **Scaling columns that has continuos data types (LotArea, GrLivArea)** 

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit_transform(merged_attributes[continuos_datatype_col])
```

```python
df_continuos_features = merged_attributes[['LotArea','GrLivArea']] 
df_continuos_features
```

```python
 df_continuos_features[continuos_datatype_col] = scaler.fit_transform(merged_attributes[continuos_datatype_col])
 df_continuos_features
```

```python
concatenating_encoded_scaled_df = pd.concat([df_continuos_features, implementing_one_hot_encode_Neighborhood], axis=1)
concatenating_encoded_scaled_df_1 = pd.concat([concatenating_encoded_scaled_df,target_attribute], axis=1)
concatenating_encoded_scaled_df_1
```

# **Training the Model using the after the selected features is processed and prepared**

```python
X = concatenating_encoded_scaled_df_1.drop(columns= target_column)  
y = concatenating_encoded_scaled_df_1[target_column]  
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression() #training the model with linear regression
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.35, random_state=50)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

# checking accuracy of our model

```python
from sklearn.metrics import mean_squared_log_error
import numpy as np
Rmsle = np.sqrt(mean_squared_log_error(y_test, y_pred)) #rmse = root mean squared error
print ("Root Mean Squared Error:", Rmsle * 100) #using rmse metric to check how good our model has performed
```
