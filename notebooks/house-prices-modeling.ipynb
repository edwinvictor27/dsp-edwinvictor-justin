## **DATA PRE-PROCESSING for the purpose of training the model**

```python
selected_features = ['LotArea', 'GrLivArea', 'Neighborhood', 'HouseStyle'] # 2 continuos and 2 Categorical features
target_feature = ['SalePrice'] #Output column
```

```python
continuos_datatype_features = ['LotArea', 'GrLivArea']
discrete_datatype_features = ['Neighborhood', 'HouseStyle']
```

```python
Extracted_Selected_Features = Training_dataset[selected_features]
Extracted_Target_Attribute = Training_dataset[target_feature]
```

```python
Extracted_Selected_Features.isnull().sum() #checking if there's any null value
```

```python
Extracted_Target_Attribute.isnull().sum()
```

```python
Merged_Features = pd.concat([Extracted_Selected_Features,Extracted_Target_Attribute], axis=1)
display(Merged_Features) #concatenating selected features together as single dataframe
```

```python
Merged_Features[discrete_datatype_features].nunique() #to get the count of different values from the categorical data type column
```

```python
encoder = OneHotEncoder(sparse_output= False)
encoded_discrete_features_df = encoder.fit(Merged_Features[discrete_datatype_features])
display(encoded_discrete_features_df) 

# performing one hot encoding for the selected discrete(categorical) features 

# cross-verifying the dimension matches with the above unique values [33 unique values hence 33 columns]
```

# **Scaling columns that has continuos data types (LotArea, GrLivArea)** 

```python
scaler = StandardScaler()
```

```python
scaler.fit_transform(Merged_Features[continuos_datatype_features])
```

```python
Extracted_Continuos_Datatype_Features = Merged_Features[continuos_datatype_features]
scaled_continuous_features = scaler.fit_transform(Extracted_Continuos_Datatype_Features)
scaled_continuous_features_df = pd.DataFrame(scaled_continuous_features, columns= continuos_datatype_features)
display(scaled_continuous_features_df)
```

```python
Transforming_encoded_features = encoder.transform(Merged_Features[discrete_datatype_features])
encoded_discrete_features_training_df = pd.DataFrame(Transforming_encoded_features, columns=encoder.get_feature_names_out(discrete_datatype_features))
encoded_discrete_features_training_df
```

```python
processed_df_1 = pd.concat([scaled_continuous_features_df , encoded_discrete_features_training_df ], axis=1)
processed_df = pd.concat([processed_df_1,Extracted_Target_Attribute], axis=1)
display(processed_df.head(5))
display(processed_df.shape)
```

# **Training the Model using the after the selected features is processed and prepared**

```python
X = processed_df.drop(columns= target_feature )  
y = processed_df[target_feature]  
```

```python
model = LinearRegression() #training the model with linear regression
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.35, random_state=50)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

# checking accuracy of our model

```python
Rmsle = np.sqrt(mean_squared_log_error(y_test, y_pred)) #rmse = root mean squared error
print ("Root Mean Squared Error:", Rmsle * 100) #using rmse metric to check how good our model has performed
```
